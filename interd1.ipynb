Conditional statements

```python
#if condition
price_tv=int(input('Enter the price of TV: '))
if price_tv<=20000:
  print("Pack it")
```

```python
#if else condition
price_tv=int(input('Enter the price of TV: '))
if price_tv<=20000:
  print("Pack it")
else:
  print("Don't pack it")
```

```python
#if elif else
#biggest of 3 number
a=float(input('Enter 1st no: '))
b=float(input('Enter 2nd no: '))
c=float(input('Enter 3rd no: '))
if a>b and a>c:
  print(a,' is big')
elif b>a and b>c:
  print(b," is big")
else:
  print(c," is big")
```

```python
#break
i=0
while i<10:
  if i==5:
    break
  print(i)
  i+=1
```

```python
#continue
i=0
while i<10:
  if i==5:
    i+=1
    continue
  print(i)
  i+=1
```
