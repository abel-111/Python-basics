```python
def is_even(number):
    if number % 2 == 0:
        return True
    else:
        return False
```

```python
a=is_even(10)
print(a)
```

3. question

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result = result * i
    return result
```

```python
a=factorial(5)
print(a)
```

4. question

```python
def sum_list(lst):
    total = 0
    for num in lst:
        total = total + num
    return total
```

```python
a=sum_list([1,2,3,4,5])
print(a)
```

7. question

```python
def fibonacci(n):
    a = 0
    b = 1
    for i in range(n):
        print(a, end=" ")
        next_num = a + b
        a = b
        b = next_num
```

```python
a=fibonacci(10)
print(a)
```

8. question

```python
def second_largest(lst):
    # Track the largest and second largest manually
    largest = lst[0]
    second = lst[0]

    for num in lst:
        if num > largest:
            second = largest
            largest = num
        elif num > second and num != largest:
            second = num

    return second
```

```python
a=second_largest([10,20,30,40,50])
print(a)
```

9. question

```python
def count_vowels(s):
    count = 0
    vowels = "aeiouAEIOU"
    for char in s:
        if char in vowels:
            count = count + 1
    return count
```

```python
a=count_vowels('abEl')
print(a)
```
